# The numerical error depends on the discretization resolution of the independent variable

The error scales with the size of the time interval 

```python
import numpy as np
import matplotlib.pyplot as plt

#------------------------------------------------------------------------------
# Numerical solution 
#------------------------------------------------------------------------------
# Parameters
k = 0.5          # decay rate constant [mM.s-1]
Ca0 = 1.0        # initial Ca concentration [mM]
T = 10.0         # total simulation time [s]


intervals = [1,0.1,0.01,0.001,0.0001,0.00001]

residuals = []

for dt in intervals:
    # Time vector
    t = np.arange(0, T + dt, dt)
    
    # Allocate array of zeros
    Ca = np.zeros_like(t)
    Ca[0] = Ca0
    
    # Euler integration loop
    for i in range(1, len(t)):
        dCa_dt = -k * Ca[i-1]
        Ca[i] = Ca[i-1] + dt * dCa_dt


    Ca_ana = Ca0 * np.exp(-k*t)
    residuals.append(np.max(np.abs(Ca - Ca_ana)))

fig, ax = plt.subplots()
ax.bar(range(len(residuals)), residuals)
ax.set_xticks(range(len(intervals)))
ax.set_xticklabels(intervals)
plt.xlabel("bin size (s)")
plt.ylabel("max error magnitude")
plt.title("Residuals")
plt.show()
    
```

## Solution for Cell number 1

```python
import numpy as np
import matplotlib.pyplot as plt

#------------------------------------------------------------------------------
# Numerical solution 
#------------------------------------------------------------------------------
# Parameters
k = 0.5          # decay rate constant [mM.s-1]
Ca0 = 1.0        # initial Ca concentration [mM]
dt = 0.01        # time step [s]
T = 10.0         # total simulation time [s]

# Time vector
t = np.arange(0, T + dt, dt)

# Allocate array of zeros
Ca = np.zeros_like(t)
Ca[0] = Ca0

# Euler integration loop
for i in range(1, len(t)):
    dCa_dt = -k * Ca[i-1]
    Ca[i] = Ca[i-1] + dt * dCa_dt




#------------------------------------------------------------------------------
# analytical solution 
#------------------------------------------------------------------------------

Ca_ana = Ca0 * np.exp(-k*t)

# Plot
plt.figure(figsize=(6,4))
plt.plot(t, Ca, linewidth=2)
plt.xlabel("Time (s)")
plt.ylabel("Ca concentration (a.u.)")
plt.title("Euler Integration of dCa/dt = -kÂ·Ca")
plt.grid(True)
plt.show()


plt.figure(figsize=(6, 4))
plt.plot(t, Ca, linewidth=2, label="Euler solution")
plt.plot(t, Ca_ana, linewidth=2, linestyle="--", alpha=0.7,
         label="Analytical solution")
plt.xlabel("Time (s)")
plt.ylabel("Ca concentration (a.u.)")
plt.title("Euler Integration of dCa/dt = -kÂ·Ca")
plt.grid(True)
plt.legend()
plt.show()


# residuals
plt.figure(figsize=(6,4))
plt.plot(t, Ca_ana-Ca, linewidth=2)
#plt.scatter(t, Ca_ana, linewidth=2)
plt.xlabel("Time (s)")
plt.ylabel("Ca concentration (a.u.)")
plt.title("Residuals")
plt.grid(True)
plt.show()
```
